## AI Adoption in U.S. Small and Medium Enterprises (2019–2024)

---

### Project Overview

**Business Problem**  
Small and medium sized enterprises (SMEs) represent 99.9% of all U.S. businesses and employ nearly half the private sector workforce. Despite the rapid democratization of AI tools especially after the launch of ChatGPT in late 2022 AI adoption among SMEs remains uneven and poorly understood. This project investigates *who* is adopting AI, *when*, *why*, and *what impact* it has on productivity and revenue growth.

**Research Questions**
1. How has AI adoption evolved among U.S. SMEs between 2019 and 2024?
2. Which industries and states lead in adoption, and which lag behind?
3. What organizational factors (size, cloud readiness, digital maturity, CEO background) predict AI adoption?
4. Does AI adoption translate into measurable productivity and revenue gains?
5. What are the primary barriers preventing SMEs from adopting AI?

**Dataset Overview**  
The analysis uses three complementary datasets spanning 2019–2024:
- **Main dataset**: 5,000 company year observations across 50 U.S. states and 10+ industries
- **Time-series dataset**: Annual adoption and investment metrics broken down by industry (60 rows)
- **National dataset**: Year-level aggregate statistics for the full SME population (6 rows)

**Project Goals**
- Understand the macro trend of AI adoption in U.S. SMEs from 2019 to 2024
- Identify key predictors of AI adoption using statistical and machine learning methods
- Quantify the productivity and revenue impact of AI investment
- Produce actionable insights for SME owners, policymakers, and technology vendors

## 1. Data Sources

The dataset was constructed by combining and synthesizing data from the following authoritative sources:

| Source | Organization | Description | Coverage |
|--------|-------------|-------------|----------|
| **Annual Business Survey (ABS)** | U.S. Census Bureau | Provides business characteristics, innovation, and technology adoption metrics for U.S. employer firms. Includes breakdowns by industry, geography, and firm size. | National, 2017–present |
| **Global AI Survey** | McKinsey & Company | Annual survey of executives across industries and geographies on AI adoption, investment, and perceived business impact. SME-specific cuts extracted. | Global, 2017–present |
| **AI Index Report** | Stanford HAI | Comprehensive annual report tracking AI development, adoption, and economic impact across sectors and countries, including SME-specific analyses. | Global, 2018–present |
| **Occupational Employment Statistics** | U.S. Bureau of Labor Statistics (BLS) | Tracks employment by occupation and industry, used to derive technology workforce percentages and wage data for SMEs. | National, annual |
| **Small Business Economic Profiles** | U.S. Small Business Administration (SBA) | State- and industry-level profiles of small business activity, revenue, employment, and technology investment patterns. | National & state-level, annual |

## 3. Data Sources Summary

```python
# ── Summary statistics per year ──────────────────────────────────────────────
summary_by_year = (
    df_main
    .groupby('year')
    .agg(
        n_companies       = ('company_id', 'count'),
        adoption_rate     = ('ai_adopted', 'mean'),
        avg_revenue_m_usd = ('revenue_m_usd', 'mean'),
        avg_employees     = ('employees', 'mean'),
        avg_ai_invest_usd = ('ai_investment_usd', 'mean'),
        pct_cloud         = ('cloud_adopted', 'mean'),
        pct_data_strategy = ('has_data_strategy', 'mean'),
    )
    .reset_index()
)

# Format for display
summary_by_year['adoption_rate']     = (summary_by_year['adoption_rate'] * 100).round(1).astype(str) + '%'
summary_by_year['pct_cloud']         = (summary_by_year['pct_cloud'] * 100).round(1).astype(str) + '%'
summary_by_year['pct_data_strategy'] = (summary_by_year['pct_data_strategy'] * 100).round(1).astype(str) + '%'

print('Year-by-Year Summary:')
summary_by_year
```

## 5. Variable Dictionary

Complete description of all columns in the main dataset (`sme_ai_adoption_main.csv`):

| Column | Type | Description |
|--------|------|-------------|
| `company_id` | string | Unique anonymized company identifier (e.g., SME_00001) |
| `year` | int | Survey year (2019–2024) |
| `state` | string | U.S. state where the company is headquartered |
| `industry` | string | Industry classification (e.g., Technology & Software, Retail & E-commerce) |
| `size_category` | string | Company size tier: Micro (1–9), Small (10–49), Medium (50–249), Upper-Mid (250–500) |
| `employees` | int | Number of full-time employees at time of survey |
| `revenue_m_usd` | float | Annual revenue in millions of USD |
| `tech_invest_pct` | float | Technology investment as a percentage of total revenue (0–1 scale) |
| `tech_invest_usd` | float | Technology investment in USD |
| `tech_emp_pct` | float | Percentage of employees in technology-related roles (0–1 scale) |
| `years_in_business` | int | Number of years the company has been operating |
| `founded_year` | int | Year the company was founded |
| `cloud_adopted` | int | Binary flag: 1 if company uses cloud services, 0 otherwise |
| `has_data_strategy` | int | Binary flag: 1 if company has a formal data strategy, 0 otherwise |
| `ceo_tech_background` | int | Binary flag: 1 if CEO/founder has a technology background, 0 otherwise |
| `digital_maturity` | float | Digital maturity score on a 1–10 scale (higher = more mature) |
| `ai_adopted` | int | Binary target variable: 1 if company has adopted AI tools, 0 otherwise |
| `ai_tools_count` | int | Number of distinct AI tools/applications in use (0 if not adopted) |
| `ai_investment_usd` | float | Annual AI-specific investment in USD (0 if not adopted) |
| `productivity_gain_pct` | float | Self-reported productivity gain attributed to technology/AI (%) |
| `revenue_growth_pct` | float | Year-over-year revenue growth rate (%) |
| `primary_barrier` | string | Main barrier preventing AI adoption (e.g., Cost, Lack of expertise, Data quality) |
